## Merubah ke lowercase, menghapus karakter, menghilangkan kata alay

```python
def lowercase(text):
    return text.lower()

def remove_char(text):
    text = re.sub('\n',' ',text) # Remove every '\n'
    text = re.sub('rt',' ',text) # Remove every retweet symbol
    text = re.sub('user',' ',text) # Remove every username
    text = re.sub('((www\.[^\s]+)|(https?://[^\s]+)|(http?://[^\s]+))',' ',text) # Remove every URL
    text = re.sub('  +', ' ', text) # Remove extra spaces
    text = re.sub('[^0-9a-zA-Z]+', ' ', text)
    text = re.sub(r'\b\d+\b', '', text)
    text = text.encode('ascii', 'replace').decode('ascii')
    text = str(text).replace('\\t'," ").replace('\\n'," ").replace('\\u'," ").replace('\\',"")
    return text

def remove_unicode(text):
#     text = re.sub(r'\bx[a-fA-F0-9]{2}\b', '', text) 
#     text = re.sub(r'\bx([a-fA-F0-9]{2})', '', text)
    text = re.sub(r'[\\x]+[a-z0-9]{2}','', text)
    return text

alay_dict_map = dict(zip(alay_dict[0], alay_dict[1]))
def normalize_alay(text):
    return ' '.join([alay_dict_map[word] if word in alay_dict_map else word for word in text.split(' ')])

print('done')
```

## Menjalankan Fungsi yang di Definisikan

```python
def preprocess(text):
    text = lowercase(text)
    text = remove_char(text)
    text = remove_unicode(text)
    text = normalize_alay(text)
    return text
print('done')
```

```python
data['Tweet'] = data['Tweet'].apply(preprocess)
print('done')
```
